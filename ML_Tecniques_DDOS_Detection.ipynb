##3.2. Limpiar y preparar los datos


```python
def eliminarColumnasVacias(df):
  ''' columnas_vacias = df.isnull().all()
  return df.drop(columnas_vacias[columnas_vacias == True].index, axis=1) '''
  valores_unicos = df.nunique()
  columnas_unicas = df.loc[:, valores_unicos == 1]
  return df.drop(columnas_unicas, axis=1)
```

```python
def missingValues(df):
  #Comprobar si hay valores vacíos
  return df.isnull().any().any()
```

```python
def fillMissingValues(df):
  #Rellenamos valores vacíos con la media
  df.fillna(df.mean(), inplace=True)
  return df
```

```python
def dibujarMatrizCorrelación(df):
    # Calcular la matriz de correlación
    corr = df.corr()
    
    plt.figure(figsize=(40, 40))

    # Crear el mapa de calor de correlación
    sns.heatmap(corr, annot=True)
    
    # Mostrar el gráfico
    plt.show()
    plt.savefig("correlation_matrix.svg", format='svg')
```

```python
def eliminarColumnasRedundantes(df):
  # Calcular la matriz de correlación
  corr_matrix = df.corr().abs()

  # Seleccionar la diagonal superior de la matriz de correlación
  upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

  # Encontrar columnas con una correlación mayor a un umbral
  to_drop = [column for column in upper.columns if any(upper[column] > 0.80)]

  # Elimina las columnas seleccionadas
  return df.drop(df[to_drop], axis=1)
```

```python
def codificarVariablesCategoricas(df):
 # Itera sobre las columnas del dataframe
  for col in df.columns:
    # Si la columna es de tipo objeto, la codifica
    if (df[col].dtype == 'object'):
      df[col] = pd.factorize(df[col])[0]
  return df
```

```python
def hayValoresExtremos(df):
  mask = np.isinf(df) | np.isnan(df) | (df > 1e5)
  return mask.sum().sum() > 0
```

```python
def tratarValoresExtremos(df):
  #Valores muy grandes o infinitos se rellenan con la media  
  df.replace(np.inf, np.nan, inplace=True)
  df.replace(np.NINF, np.nan, inplace=True)
  mean = df.mean()
  df.fillna(mean, inplace=True)

  return df
```

```python
def normalizarDatos(df):
  scaler = MinMaxScaler()
  scaled_df = scaler.fit_transform(df)
  return pd.DataFrame(scaled_df, columns=df.columns)
                      
  ''' return (df - df.mean()) / df.std() '''
```

```python
def analisisPCA(df):
  explained_pca_ratio = 0.90
  #print(df.columns)
  data_scaled = pd.DataFrame(preprocessing.scale(df), columns=df.columns)

  pca = PCA().fit(data_scaled)
  line_data = np.cumsum(pca.explained_variance_ratio_)
  line_data = np.insert(line_data, 0, 0)
  plt.bar(np.arange(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_, color='g')
  plt.plot(np.arange(0, len(line_data)), line_data, marker='D')
  plt.xlim(0, len(pca.explained_variance_ratio_), 1)
  plt.axhline(y=explained_pca_ratio, color='black', linestyle='--')
  plt.xlabel('Number of components')
  plt.ylabel('Cumulative explained variance')

  sklearn_pca = PCA(n_components=explained_pca_ratio)
  return sklearn_pca.fit_transform(data_scaled)
  #print(data)
```

```python
def prepararDatos(df):
  
  df = df.drop(columns=[df.columns[0], "Flow ID", "Timestamp"], axis=1)
  #print(df.columns)
  print("Eliminando columnas vacías")
  df = eliminarColumnasVacias(df)

  print("Rellenando valores vacíos")
  if missingValues(df):
    fillMissingValues(df)

  dibujarMatrizCorrelación(df)
  
  print("Eliminando columnas redundantes")
  df = eliminarColumnasRedundantes(df)
  
  print("Número de columnas resultantes: ",df.shape[1],"\n")
  print(df.columns)
  #dibujarMatrizCorrelación(df)

  #Codificar variables categóricas
  df = codificarVariablesCategoricas(df)

  #Tratar valores extremos y rellenarlos con la media
  #print(hayValoresExtremos(df))
  df = tratarValoresExtremos(df)

  #Normalizar datos
  df = normalizarDatos(df)

  #PCA (reducción columnas) Comparar su uso y no uso (pendiente)
  #df = analisisPCA(df) 
  #df = kernelPca(df)

  #Dividir los datos en conjunto de entrenamiento y prueba:
  X = df.drop(['Label'], axis=1)
  y = df['Label'] 
  return train_test_split(X, y, test_size=0.3)
```

##3.3. Seleccionar algoritmo

```python
def clasificadorAleatorio():
  return DummyClassifier(strategy="uniform")
```

```python
def regresionLogistica():
  return LogisticRegression(penalty='l1', C=0.83, solver='saga', multi_class='multinomial', max_iter=200)
  #solver='saga', para grandes conjuntos de datos, soporta penalty='l1', lbfgs no (es el por defecto)
  #C=0.83, por defecto 1, con un numero mayor o menor, empeora el resultado
  #penalty='l1', para reducir la dimensionalidad de los datos, ya que tenemos muchas características
  #multi_class='multinomial', para tener en cuenta las correlaciones entre clases
  #max_iter=200, mejora la precisión, por defecto 100
```

```python
def knn():
  return KNeighborsClassifier(n_neighbors=35, algorithm='kd_tree', leaf_size=20, p=1)
  #n_neighbors, se suele poner de 1-10 para datasets pequeños o medianos, más si son más grandes
  #weights: si el dataset no fuese balanced, convendría utilizar distance en lugar de uniform que es el por defecto
  #algorithm: kd_tree o ball_tree para datasets grandes
  #leaf_size: para kd_tree o ball_tree
  #p: distancia, se suele utilizar 2. Puede usarse 1 (euclidea) o inf (Chebyshev)
```

```python
def svm():
  return SVC(C=2, kernel='linear', decision_function_shape='ovo', tol=1e-7)
  #C Un valor más alto de C indica que se permite un mayor grado de sobreajuste para maximizar la precisión. valor más bajo de C más peso a la generalización para obtener una mejor precisión en los datos no vistos.
  #kernel: lineal, poly y rbf
  #decision_function_shape: ovr por defecto se ajusta un clasificador SVM para cada clase y se elige la clase con el mayor valor de la función de decisión. ovo para ajustar clasificadores SVM para cada par de clases.
  #tol (precisión del algoritmo) Valores más bajos de tol dan lugar a una solución más precisa pero aumentan el tiempo de ejecución
```

```python
def randomForest():
  return RandomForestClassifier(criterion='entropy', max_depth=10, max_features=10)
  #criterion='entropy', por defecto gini, al cambiarlo a entropy, se mejora la precisión
  #max_depth=13, por defecto None, al aumentarlo, se mejora la precisión
  #max_features=10, por defecto todas, al disminuirlo se aumenta la precisión, si se aumenta, la precisión empeora
```

##3.4. Entrenar el modelo con el conjunto de entrenamiento:


```python
def entrenamiento (model, X_train, y_train):
  model.fit(X_train, y_train)
  return model
```

##3.5. Evaluar el rendimiento del modelo con el conjunto de prueba

```python
def evaluarRendimiento (model, X_test, y_test):
  #Hacer predicciones con el conjunto de prueba
  y_pred = model.predict(X_test)
  #Calcular la precisión del modelo
  #accuracy = accuracy_score(y_test, y_pred)
  precision = precision_score(y_test, y_pred)
  recall = recall_score(y_test, y_pred)
  conf_matrix = confusion_matrix(y_test, y_pred)
  f1 = f1_score(y_test, y_pred)

  # Imprimir las métricas de rendimiento
  #print("Accuracy:", accuracy)
  print("Precision:", precision)
  print("Recall:", recall)
  print("F1-score:", f1)
  conf_matrix_list = np.ndarray.tolist(conf_matrix)
  table = PrettyTable()
  table.field_names = ["", "Predicción negativa", "Predicción positiva"]
  table.add_row(["Real negativa", conf_matrix_list[0][0], conf_matrix_list[0][1]])
  table.add_row(["Real positiva", conf_matrix_list[1][0], conf_matrix_list[1][1]])
  print("Confusion Matrix:")
  print(table)  
```
